```python
df_costumers = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load("/Volumes/consumer_goods/files/input_data_files/parent_data/dim_customers.csv")
df_costumers.printSchema()
```

```python
df_costumers.show(10)
```

Customer code column should be string as it is not a agg column

```python
import pyspark.sql.functions as F
df_costumers = df_costumers.withColumn("customer_code", F.col("customer_code").cast("string") )
df_costumers.printSchema()
```

```python
df_costumers.show(10)
```

Create table in gold schema

```python
df_costumers.write.mode("overwrite").saveAsTable("consumer_goods.gold.dim_customers")
```

```python
%sql
select * from consumer_goods.gold.dim_customers limit 10;
```
