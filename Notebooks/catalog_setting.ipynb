Create main catalog for the company

```sql
create catalog if not exists consumer_goods;
```

```sql
show catalogs;
```

Our catalog is available

```sql
use catalog consumer_goods;
```

Create gold schema for the parent company

```sql
create schema if not exists consumer_goods.gold;
```

```sql
show schemas;
```

Create bronze and silver schemas for the child companies

```sql
create schema if not exists consumer_goods.silver;
create schema if not exists consumer_goods.bronze;
```

```sql
show schemas;
```
